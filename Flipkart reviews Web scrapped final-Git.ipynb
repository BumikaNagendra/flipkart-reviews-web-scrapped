### <font color=orange> Problem Statement </font> :  

Analysing Customer's choice for Flipkart based on the reviews and ratings. 

Evaluation of an individual's sentiment from a piece of text is challenging.

Sentiment analysis is necessary to get customer feelings,attitudes, reactions, and opinions of existing and potential buyers towards flipkart's products or services.

So lets apply the NLP techniques to analyze the customer reviews in Flipkart.

# Flipkart Electronics reviews Web scrapped Data

##### Importing required libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline
```

### Product 1 : wired headset

```python
df=pd.DataFrame()
for i in range(1,520):
    url='https://www.flipkart.com/redmi-mi-hi-resolution-audio-wired-headset/product-reviews/itmfc704fbf24f2c?pid=ACCFUT9HGEZ5XNKZ&lid=LSTACCFUT9HGEZ5XNKZLFX2ZK&marketplace=FLIPKART&page={}'.format(i)
    category=url.split('/')[3]
    HEADERS = ({'User-Agent':
            'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
            'Accept-Language': 'en-US, en;q=0.5'})
    
    data = requests.get(url,headers=HEADERS)
    soup = BeautifulSoup(data.content)
    category=url.split('/')[3]
    parent_tags=soup.findAll('div',{'class':'col _2wzgFH K0kLPL'})
    
    for parent in parent_tags:
        try:
            review_headline=parent.find('p',{'class':'_2-N8zT'}).text
        except:
            review_headline=None
        try:
            review=parent.find('div',{'class':'t-ZTKy'}).text
        except:
            review=None
        try:
            rate=parent.select_one('div',{'class':'_3LWZlK _1BLPMq','class':'_3LWZlK _1rdVr6 _1BLPMq'}).text
                
        except:
            rate=None
    
        temp={
        'review_title':review_headline,
        'review':review,
        'rate':rate,
        'category':category}
        
        df=df.append(temp, ignore_index=True)
        
df[['review_title','review','rate','category']]
df['rating']=df['rate'].str.extract('(\d+)')

df.head()
    
```

```python
df.shape
```

```python
df.head()
```
